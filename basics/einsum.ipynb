# Einsum Operator

# 1. Import Dependencies

```python
import numpy as np
```

```python
A = np.array([
    [2, 6, 5, 2],
    [2, -2, 2, 3],
    [1, 5, 4, 0],
])

B = np.array([
    [2, 9, 0, 3, 0],
    [3, 6, 8, -2, 2],
    [1, 3, 5, 0, 1],
    [3, 0, 2, 0, 5]
])

A.shape, B.shape
```

```python
np.matmul(A, B)
```

```python
np.einsum('ij,jk -> ik', A,B)
```

```python
C = np.array([
    [2, 6, 5, 2],
    [2, -2, 2, 3],
    [1, 5, 4, 0],
])

D = np.array([
    [2, 9, 0, 3],
    [3, 6, 8, -2],
    [1, 3, 5, 0],
])

print("Hardamond C := \n")
print(C*D, "\n")

print("Einsum C =: \n")
print(np.einsum('ij, ij -> ij', C,D))
```

```python
print("Transpose A =: \n")
print(A.T, "\n")

print("Einsum Transpose A =: \n")
print(np.einsum("ij -> ji", A))
```

```python
E = np.array([
   [[2, 6, 5, 2],
    [2, -2, 2, 3],
    [1, 5, 4, 0]],
   [[1, 3, 1, 22],
    [0, 2, 2, 0],
    [1, 5, 4, 1]]
])

F = np.array([
   [[2, 9, 0, 3, 0],
    [3, 6, 8, -2, 2],
    [1, 3, 5, 0, 1],
    [3, 0, 2, 0, 5]],
   
   [[1, 0, 0, 3, 0],
    [3, 0, 4, -2, 2],
    [1, 0, 2, 0, 0],
    [3, 0, 1, 1, 0]]
])

E.shape, F.shape
```

```python
print("Batch Multiplication G := \n")
print(np.matmul(E,F), "\n")
```

```python
print("Einsum G := \n")
print(np.einsum('bij, bjk -> bik', E, F))
```

```python
A = np.array([
    [2, 6, 5, 2],
    [2, -2, 2, 3],
    [1, 5, 4, 0],
])

print('Axix 0 Sum A =: \n')
print(np.sum(A, axis=0), "\n")

print('Einsum A =: \n')
print(np.einsum("ij -> j", A))
print("------------------------------")

print('Axix 1 Sum A =: \n')
print(np.sum(A, axis=1), "\n")

print('Einsum A =: \n')
print(np.einsum("ij -> i", A))
```

```python
# Q = batchsize, sequence_query, modelsize
# K = batch_size, sequence_key, modelsize

QUERY = np.random.randn(32, 64, 512) # bqm
KEY = np.random.randn(32, 128, 512) # bkm
```
